## Demo: How to work with database


Download database file.


```python
#!wget https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBM-DA0321EN-SkillsNetwork/LargeData/m4_survey_data.sqlite
import requests

url = 'https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBM-DA0321EN-SkillsNetwork/LargeData/m4_survey_data.sqlite'
response = requests.get(url)

with open('m4_survey_data.sqlite', 'wb') as f:
    f.write(response.content)
```

Connect to the database.


```python
import sqlite3
conn = sqlite3.connect("m4_survey_data.sqlite") # open a database connection
```

Import pandas module.


```python
import pandas as pd
import matplotlib.pyplot as plt
```

## Demo: How to run an sql query


```python
# print how many rows are there in the table named 'master'
QUERY = """
SELECT COUNT(*)
FROM master
"""

# the read_sql_query runs the sql query and returns the data as a dataframe
df = pd.read_sql_query(QUERY,conn)
df.head()
```

## Demo: How to list all tables


```python
# print all the tables names in the database
QUERY = """
SELECT name as Table_Name FROM
sqlite_master WHERE
type = 'table'
"""
# the read_sql_query runs the sql query and returns the data as a dataframe
pd.read_sql_query(QUERY,conn)
```

## Demo: How to run a group by query


```python
QUERY = """
SELECT Age,COUNT(*) as count
FROM master
group by age
order by age
"""
pd.read_sql_query(QUERY,conn)
```

## Demo: How to describe a table


```python
table_name = 'master'  # the table you wish to describe

QUERY = """
SELECT sql FROM sqlite_master
WHERE name= '{}'
""".format(table_name)

df = pd.read_sql_query(QUERY,conn)
print(df.iat[0,0])
```

# Hands-on Lab


## Visualizing distribution of data


### Histograms


Plot a histogram of `ConvertedComp.`


```python
QUERY = """SELECT * FROM master"""
df = pd.read_sql_query(QUERY,conn)
df['ConvertedComp'].plot(kind='hist', bins=20, xlim=(0, df['ConvertedComp'].max()), edgecolor='black')
plt.title('Histogram of ConvertedComp')
plt.ylabel('Number of Respondents')
plt.xlabel('Annual Compensation in USD')
plt.ticklabel_format(style='plain', axis='x')
plt.xticks(rotation=90)
plt.show()
```

### Box Plots


Plot a box plot of `Age.`


```python
print(df['Age'].describe())
df['Age'].plot(kind='box')
plt.title('Box Plot of Age')
plt.ylabel('Age Range')
plt.show()
```

## Visualizing relationships in data


### Scatter Plots


Create a scatter plot of `Age` and `WorkWeekHrs.`


```python
print(f"Correlation between 'Age' and 'WorkWeekHrs': {df['Age'].corr(df['WorkWeekHrs'])}")
df.plot(kind='scatter', x='Age', y='WorkWeekHrs')
plt.title('Scatter Plot of Age vs WorkWeekHrs')
plt.show()
```

### Bubble Plots


Create a bubble plot of `WorkWeekHrs` and `CodeRevHrs`, use `Age` column as bubble size.


```python
normalize_age = (df['Age'] - df['Age'].min()) / (df['Age'].max() - df['Age'].min())
df.plot(kind='scatter', x='WorkWeekHrs', y='CodeRevHrs', s= normalize_age * 500, alpha=0.5)
plt.title('Scatter Plot of Code Review Hours vs Work Week Hours with Point Sizes Indicating Normalized Age')
plt.legend(['Age'])
plt.show()
```

## Visualizing composition of data


### Pie Charts


Create a pie chart of the top 5 databases that respondents wish to learn next year. Label the pie chart with database names. Display percentages of each database on the pie chart.


```python
QUERY = """SELECT * FROM DatabaseDesireNextYear"""
df2 = pd.read_sql_query(QUERY,conn)
df2_top_five = df2.groupby('DatabaseDesireNextYear').count().sort_values(by='Respondent', ascending=False).head()
df2_top_five['Respondent'].plot(kind='pie', autopct='%1.1f%%',)
plt.title('Top 5 Databases Respondents Desire to Learn Next Year')
plt.ylabel('')
plt.show()
```

### Stacked Charts


Create a stacked chart of median `WorkWeekHrs` and `CodeRevHrs` for the age group 30 to 35.


```python
QUERY = """SELECT * FROM master WHERE Age BETWEEN 30 AND 35"""
df3 = pd.read_sql_query(QUERY,conn)
df3 = df3.groupby('Age')[['WorkWeekHrs', 'CodeRevHrs']].median()
df3.index = df3.index.astype(int)
df3.plot(kind='bar', stacked=True)
plt.title('Stacked Chart of Median WorkWeekHrs and CodeRevHrs for the age group 30 to 35')
plt.ylabel('Hours')
plt.xlabel('Age')
plt.xticks(rotation=None)
plt.legend(bbox_to_anchor=(1, 1))
plt.show()
```

## Visualizing comparison of data


### Line Chart


Plot the median `ConvertedComp` for all ages from 45 to 60.


```python
QUERY = """SELECT * FROM master WHERE Age BETWEEN 45 AND 60"""
df4 = pd.read_sql_query(QUERY,conn)
df4 = df4.groupby('Age')['ConvertedComp'].median()
df4.plot(kind='line', xlim=(df4.index.min(),df4.index.max()))
plt.title('Line Chart of Median Annual Compensation for Ages Between 45 and 60')
plt.ylabel('Annual Compensation in USD')
plt.xlabel('Age')
plt.show()
```

### Bar Chart


Create a horizontal bar chart using column `MainBranch.`


```python
mainbranch_value_count = df['MainBranch'].value_counts()
mainbranch_value_count.plot(kind='barh')
plt.title('Line Chart of Median Annual Compensation for Ages Between 45 and 60')
plt.ylabel('Annual Compensation in USD')
plt.xlabel('Coun'))
plt.show()
```
